```python
import pandas as pd
import numpy as np
from sklearn import model_selection
from sklearn import ensemble
from sklearn import svm
from sklearn import neighbors
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.ensemble import VotingClassifier
```

```python
df=pd.read_csv('ENB_data.csv',sep=';')
df.describe()
```

```python
df.corr()
```

```python
df
```

```python
df['total_charges']=df.cooling_Load+df.heating_Load
df['charges_classes']=pd.qcut(df.total_charges,q=4,labels=False)
```

```python
df['charges_classes']=pd.qcut(df.total_charges,q=4,labels=False)
```

```python
data=df.drop(['cooling_Load','heating_Load','total_charges','charges_classes'],axis=1)
X_train,X_test,y_train,y_test=train_test_split(data,df.charges_classes,test_size=0.2,random_state=34)
```

```python
scaler=preprocessing.StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
knn=neighbors.KNeighborsClassifier()
param_knn={'n_neighbors':np.arange(2,51)}
knn.fit(X_train_scaled,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
gs = GridSearchCV(knn, param_knn, cv=5)
gs.fit(X_train_scaled, y_train)
best_k = gs.best_params_['n_neighbors']


print(f"Meilleur hyperparamètre: {best_k}")
```

```python
y_pred=gs.predict(X_test_scaled)
pd.crosstab(y_test,y_pred,colnames=['classe prédite'],rownames=['classe réelle'])
```

```python
from sklearn.svm import SVC
svm=SVC()
param_svm={'kernel':['rbf','linear'],'C':[0.1,1,10,50]}
svm.fit(X_train_scaled,y_train)
```

```python
gs_svm=GridSearchCV(svm,param_svm,cv=5)
gs_svm.fit(X_train_scaled,y_train)
print("les meilleurs parametres sont : ",gs_svm.best_params_)
```

```python
y_pred=gs_svm.predict(X_test_scaled)
pd.crosstab(y_test,y_pred,colnames=['classe prédite'],rownames=['classe réelle'])
```

```python
rf=ensemble.RandomForestClassifier()
rf.fit(X_train_scaled,y_train)
rf.score(X_test_scaled,y_test)
```

```python
param_rf={'max_features':['sqrt','log2',None],'min_samples_split':range(2,31,2)}
```

```python
gs_rf=GridSearchCV(rf,param_rf,cv=5)
gs_rf.fit(X_train_scaled,y_train)
print("les meilleurs parametres sont : ",gs_rf.best_params_)
```

```python
y_pred=gs_rf.predict(X_test_scaled)

pd.crosstab(y_test,y_pred,colnames=['classes prédites'],rownames=['classes réelles'])
```
